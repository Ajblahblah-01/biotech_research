```python
# prompt: add another column called temp_name, and make the name go from a to z and then aa and so on

import string

def generate_temp_name(index):
  alphabet = string.ascii_lowercase
  if index < len(alphabet):
    return alphabet[index]
  else:
    quotient = index // len(alphabet)
    remainder = index % len(alphabet)
    return alphabet[quotient] + alphabet[remainder]

df['temp_name'] = [generate_temp_name(i) for i in range(len(df))]
df.head(100)
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

def create_driver(given_text):

    # Open a specific URL in each tab
    driver.get("http://www.swissadme.ch/index.php")  # Replace with the URL you want to open

    # Find the sample text box and submit button by their IDs or other locators
    text_box = driver.find_element(By.ID, "smiles")  # Replace "textbox_id" with actual ID

    # Write text into the sample text box
    text_to_write = given_text
    text_box.send_keys(text_to_write)

    submit_button = driver.find_element(By.ID,"submitButton")  # Replace "submit_button_id" with actual ID

    # Click the submit button
    submit_button.click()

    # Wait for a short duration to simulate human-like behavior (optional)
    # define the condition for the wait
    # submit_button_new = driver.find_element(By.ID,"submitButton")
    # while submit_button_new.get_attribute("disabled") == True:
    #     driver.implicitly_wait(1) # seconds

    element = WebDriverWait(driver, 1000000000).until(
        EC.element_to_be_clickable((By.ID, "submitButton"))
    )

    download_csv = driver.find_element(By.XPATH,'//a[contains(@href, "results")]/img[@src="/shared/img/csv-48.png"]')
    # Click on the button
    download_csv.click()
```

```python
def create_string(df,start,end):
    result_string = ""
    for i in range(start,end):
        temp_string = f"{df.iloc[i, 1]} {df.iloc[i, 2]} \n"
        result_string += temp_string
    return result_string
```

```python
import os
from tqdm.notebook import tqdm
import time

driver = gs.Chrome(options=options)

batch_size = 50


for i in tqdm(range(0,len(df),batch_size)) :
    temp_string = create_string(df,i,i+batch_size)
    # print(df.iloc[i, 1],df.iloc[i, 2])
    create_driver(temp_string)
    # # Specify the current and new file names
    current_file_name = 'swissadme.csv'
    new_file_name = f"swissadme{round(i/50)}.csv"
    time.sleep(10)
    # Rename the file
    if os.path.isfile(current_file_name):
        os.rename(current_file_name, new_file_name)

```

```python
# Format data
import pandas as pd
import glob

# Get a list of all files matching the pattern "swissadme*"
files = glob.glob("swissadme*")

# Create an empty list to store the dataframes
dataframes = []

# Loop through the list of files
for file in files:
    # print(file)
    # Read the data from the current file
    tempo_df = pd.read_csv(file)
    print(file,len(tempo_df))

    # # Append the current dataframe to the list
    dataframes.append(tempo_df)

# Concatenate all the dataframes into a single dataframe
merged_df = pd.concat(dataframes)
```

```python
# prompt: make a map with key as temp_name from df and value as compound name

name_map = dict(zip(df['temp_name'], df['Compound Name']))
print(name_map)
```

```python
# prompt: traverse merged_df and for every row check for Molecule in name_map as key and replace it with respective value

for index, row in merged_df.iterrows():
  for key, value in name_map.items():
    if key in row['Molecule']:
      merged_df.loc[index, 'Molecule'] = value
      break

# Print the modified dataframe
merged_df.head()
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

# Open the initial tab
driver.get("http://www.swissadme.ch/index.php")  # Replace with your URL


# Open multiple tabs using keyboard shortcut (Ctrl/Cmd + T)
num_tabs_to_open = 4
for _ in range(num_tabs_to_open):
    driver.find_element_by_tag_name('body').send_keys(Keys.CONTROL + 't')  # For Windows/Linux
    # driver.find_element_by_tag_name('body').send_keys(Keys.COMMAND + 't')  # For macOS

# Switch to each tab and perform actions
for tab_index in range(num_tabs_to_open + 1):  # Add 1 for the original tab
    driver.switch_to.window(driver.window_handles[tab_index])

    # Open a specific URL in each tab
    driver.get("http://www.swissadme.ch/index.php")  # Replace with the URL you want to open

    # Find the sample text box and submit button by their IDs or other locators
    text_box = driver.find_element_by_id("textbox_id")  # Replace "textbox_id" with actual ID
    submit_button = driver.find_element_by_id("submit_button_id")  # Replace "submit_button_id" with actual ID

    # Write text into the sample text box
    text_to_write = f"This is tab {tab_index + 1}"
    text_box.send_keys(text_to_write)

    # Click the submit button
    submit_button.click()

    # Wait for a short duration to simulate human-like behavior (optional)
    time.sleep(1)

# Close the WebDriver
driver.quit()
```
